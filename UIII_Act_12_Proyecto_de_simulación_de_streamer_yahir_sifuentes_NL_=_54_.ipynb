```python
# ============================================
# SISTEMA INTELIGENTE DE STREAMERS
# Programación Orientada a Objetos (POO)
# Profesor: Eliseo Nava
# Materia: Elabora Proyectos con Programación Lógica
# CBTIS 128
# Especialidad: Inteligencia Artificial (IA)
# ============================================

# ============================================
# CLASE STREAMER
# ============================================

class Streamer:

    def __init__(self, nombre, plataforma, horario,
                 setup_pc, comunidad, juegos, alertas):

        self.nombre = nombre
        self.plataforma = plataforma
        self.horario = horario
        self.setup_pc = setup_pc
        self.comunidad = comunidad
        self.juegos = juegos
        self.alertas = alertas


# ============================================
# LISTA PRINCIPAL
# ============================================

streamers = []


# ============================================
# FUNCION AGREGAR STREAMER
# ============================================

def agregar_streamer():

    print("\n===== REGISTRO DE STREAMER =====")

    nombre = input("Nombre del streamer: ")
    plataforma = input("Plataforma (Twitch/Kick): ")
    horario = input("Horario de emisión: ")
    setup_pc = input("Setup de PC: ")
    comunidad = int(input("Cantidad de subscriptores: "))
    juegos = input("Juegos principales: ")
    alertas = input("Tipo de alertas: ")

    nuevo_streamer = Streamer(
        nombre,
        plataforma,
        horario,
        setup_pc,
        comunidad,
        juegos,
        alertas
    )

    streamers.append(nuevo_streamer)

    print("\nStreamer agregado correctamente.")


# ============================================
# FUNCION LISTAR STREAMERS
# ============================================

def listar_streamers():

    print("\n================ LISTA DE STREAMERS ================\n")

    if len(streamers) == 0:

        print("No hay streamers registrados.")

    else:

        print("-" * 160)

        print(
            f"{'No.':<5}"
            f"{'NOMBRE':<20}"
            f"{'PLATAFORMA':<15}"
            f"{'HORARIO':<20}"
            f"{'SETUP PC':<30}"
            f"{'COMUNIDAD':<15}"
            f"{'JUEGOS':<30}"
            f"{'ALERTAS':<20}"
        )

        print("-" * 160)

        contador = 1

        for s in streamers:

            print(
                f"{contador:<5}"
                f"{s.nombre:<20}"
                f"{s.plataforma:<15}"
                f"{s.horario:<20}"
                f"{s.setup_pc:<30}"
                f"{s.comunidad:<15}"
                f"{s.juegos:<30}"
                f"{s.alertas:<20}"
            )

            contador += 1

        print("-" * 160)


# ============================================
# FUNCION ACTUALIZAR STREAMER
# ============================================

def actualizar_streamer():

    listar_streamers()

    if len(streamers) == 0:
        return

    numero = int(input("\nIngrese el número del streamer a actualizar: "))

    if numero >= 1 and numero <= len(streamers):

        s = streamers[numero - 1]

        print("\n===== ACTUALIZAR DATOS =====")

        s.nombre = input("Nuevo nombre: ")
        s.plataforma = input("Nueva plataforma: ")
        s.horario = input("Nuevo horario: ")
        s.setup_pc = input("Nuevo setup de PC: ")
        s.comunidad = int(input("Nueva cantidad de subscriptores: "))
        s.juegos = input("Nuevos juegos principales: ")
        s.alertas = input("Nuevo tipo de alertas: ")

        print("\nStreamer actualizado correctamente.")

    else:
        print("\nNúmero no válido.")


# ============================================
# FUNCION BORRAR STREAMER
# ============================================

def borrar_streamer():

    listar_streamers()

    if len(streamers) == 0:
        return

    numero = int(input("\nIngrese el número del streamer a borrar: "))

    if numero >= 1 and numero <= len(streamers):

        eliminado = streamers.pop(numero - 1)

        print("\nStreamer eliminado correctamente.")

    else:
        print("\nNúmero no válido.")


# ============================================
# FUNCION IA SIMPLE
# ============================================

def analisis_ia():

    print("\n========== ANALISIS INTELIGENTE ==========")

    if len(streamers) == 0:

        print("No hay streamers registrados.")

    else:

        for s in streamers:

            print("\n-----------------------------------")
            print("Streamer:", s.nombre)
            print("Plataforma:", s.plataforma)
            print("Comunidad:", s.comunidad)

            # Inteligencia artificial simple
            if s.comunidad >= 10000:
                print("IA: Streamer muy popular.")

            elif s.comunidad >= 5000:
                print("IA: Streamer en crecimiento.")

            else:
                print("IA: Streamer nuevo en la plataforma.")

            if s.plataforma.lower() == "twitch":
                print("IA: Excelente para directos gaming.")

            elif s.plataforma.lower() == "kick":
                print("IA: Plataforma moderna y competitiva.")

            if s.alertas.lower() == "animadas":
                print("IA: Usa alertas llamativas para atraer audiencia.")

            else:
                print("IA: Puede mejorar sus alertas visuales.")


# ============================================
# MENU PRINCIPAL
# ============================================

def menu():

    while True:

        print("\n===================================")
        print("   SISTEMA INTELIGENTE STREAMER")
        print("===================================")

        print("1. Agregar streamer")
        print("2. Listar streamers")
        print("3. Actualizar streamer")
        print("4. Borrar streamer")
        print("5. Analisis IA")
        print("6. Salir")

        opcion = input("\nSeleccione una opción: ")

        # ====================================
        # CONDICIONES DEL MENU
        # ====================================

        if opcion == "1":
            agregar_streamer()

        elif opcion == "2":
            listar_streamers()

        elif opcion == "3":
            actualizar_streamer()

        elif opcion == "4":
            borrar_streamer()

        elif opcion == "5":
            analisis_ia()

        elif opcion == "6":

            print("\nPrograma realizado por Yahir Sifuentes")
            print("CBTIS 128 - Inteligencia Artificial")

            break

        else:
            print("\nOpción no válida.")


# ============================================
# INICIO DEL PROGRAMA
# ============================================

menu()


```
