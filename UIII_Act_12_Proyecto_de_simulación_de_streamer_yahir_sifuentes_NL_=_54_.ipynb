```python
# ============================================
# SISTEMA INTELIGENTE DE STREAMERS
# Adaptado para Clase Streamer
# Profesor: Eliseo Nava
# Materia: Elabora Proyectos Con Programacion Logica
# CBTIS 128
# Especialidad: Inteligencia Artificial (IA)
# ============================================

# Lista de diccionarios (inicia vacía)
streamers = []

# ============================================
# FUNCION PARA AGREGAR STREAMERS
# ============================================
def agregar_streamer():
    global streamers

    nuevo_streamer = {}

    print("\n===== REGISTRO DE STREAMER =====")

    nuevo_streamer["nombre"] = input("Nombre del streamer: ")
    nuevo_streamer["plataforma"] = input("Plataforma (Twitch/Kick): ")
    nuevo_streamer["horario"] = input("Horario de emisión: ")
    nuevo_streamer["setup_pc"] = input("Setup de PC: ")
    nuevo_streamer["subscriptores"] = input("Cantidad de subscriptores: ")
    nuevo_streamer["juegos"] = input("Juegos principales: ")
    nuevo_streamer["alertas"] = input("Tipo de alertas: ")

    streamers.append(nuevo_streamer)

    print("\nStreamer agregado correctamente.\n")

# ============================================
# FUNCION PARA MOSTRAR STREAMERS
# ============================================
def mostrar_streamers():

    print("\n============= LISTADO DE STREAMERS =============")

    if len(streamers) == 0:
        print("No hay streamers registrados.\n")

    else:

        print("-" * 140)

        print(
            f"{'NOMBRE':<15}"
            f"{'PLATAFORMA':<15}"
            f"{'HORARIO':<20}"
            f"{'SETUP':<25}"
            f"{'SUBS':<15}"
            f"{'JUEGOS':<25}"
            f"{'ALERTAS':<20}"
        )

        print("-" * 140)

        for s in streamers:

            print(
                f"{s['nombre']:<15}"
                f"{s['plataforma']:<15}"
                f"{s['horario']:<20}"
                f"{s['setup_pc']:<25}"
                f"{s['subscriptores']:<15}"
                f"{s['juegos']:<25}"
                f"{s['alertas']:<20}"
            )

        print("-" * 140)

# ============================================
# FUNCION PARA BUSCAR STREAMER
# ============================================
def buscar_streamer():

    buscar = input("\nIngrese el nombre del streamer a buscar: ")

    encontrado = False

    for s in streamers:

        if s["nombre"].lower() == buscar.lower():

            print("\nSTREAMER ENCONTRADO")
            print("---------------------------")

            print("Nombre:", s["nombre"])
            print("Plataforma:", s["plataforma"])
            print("Horario:", s["horario"])
            print("Setup PC:", s["setup_pc"])
            print("Subscriptores:", s["subscriptores"])
            print("Juegos principales:", s["juegos"])
            print("Alertas:", s["alertas"])

            encontrado = True

    if encontrado == False:
        print("\nNo se encontró un streamer con ese nombre.")

# ============================================
# FUNCION PARA ELIMINAR STREAMER
# ============================================
def eliminar_streamer():

    eliminar = input("\nIngrese el nombre del streamer a eliminar: ")

    encontrado = False

    for s in list(streamers):

        if s["nombre"].lower() == eliminar.lower():

            streamers.remove(s)

            print("\nStreamer eliminado correctamente.")

            encontrado = True

            break

    if not encontrado:
        print("\nNo existe un streamer con ese nombre.")

# ============================================
# FUNCION DE ANALISIS INTELIGENTE
# ============================================
def analisis_inteligente():

    print("\n===== ANALISIS INTELIGENTE =====")

    if len(streamers) == 0:

        print("No hay streamers registrados.")

    else:

        for s in streamers:

            print("\nStreamer:", s["nombre"])

            # ANALISIS DE SUBSCRIPTORES
            if int(s["subscriptores"]) >= 10000:
                print("Streamer muy popular.")
            elif int(s["subscriptores"]) >= 5000:
                print("Streamer en crecimiento.")
            else:
                print("Streamer nuevo en la plataforma.")

            # ANALISIS DE PLATAFORMA
            if s["plataforma"].lower() == "twitch":
                print("Streamer activo en Twitch.")
            elif s["plataforma"].lower() == "kick":
                print("Streamer activo en Kick.")
            else:
                print("Plataforma registrada:", s["plataforma"])

            # ANALISIS DE ALERTAS
            if s["alertas"].lower() == "animadas":
                print("Usa alertas llamativas para la comunidad.")
            else:
                print("Tipo de alertas:", s["alertas"])

# ============================================
# MENU PRINCIPAL
# ============================================
while True:

    print("\n===================================")
    print("     SISTEMA INTELIGENTE STREAMER")
    print("===================================")

    print("1. Agregar streamer")
    print("2. Mostrar streamers")
    print("3. Buscar streamer")
    print("4. Eliminar streamer")
    print("5. Analisis inteligente")
    print("6. Salir")

    opcion = input("\nSeleccione una opción: ")

    # ========================================
    # CONDICIONES
    # ========================================

    if opcion == "1":
        agregar_streamer()

    elif opcion == "2":
        mostrar_streamers()

    elif opcion == "3":
        buscar_streamer()

    elif opcion == "4":
        eliminar_streamer()

    elif opcion == "5":
        analisis_inteligente()

    elif opcion == "6":

        print("programa realizado por yahir sifuentes  NL = 54  ")
        print("Programa adaptado para Clase Streamer")

        break

    else:
        print("\nOpción no válida.")
```
